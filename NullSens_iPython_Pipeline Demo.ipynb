### FASTA Formatting

```python
import re

seqFile = 'formatSeqs/rousk_ph_gradient_seqs.fna'
inputFile = open(seqFile, "r")
outputFile = open(seqFile.partition('.')[0] + '_formated.fasta',"w")

for line in inputFile:
    x = re.search('[ACTG]{20}',str(line))
    tempLine = line.partition(str(x.group(0)))
    outputFile.write(str(tempLine[0])+'\n'+str(x.group(0))+tempLine[2])

inputFile.close()
outputFile.close()
```

```python
%%bash
head -n 10 formatSeqs/rousk_ph_gradient_seqs_formated.fasta
```

### Clean FASTA Headers 

```python
seqFile = seqFile.partition('.')[0] + '_formated.fasta'
inputList = open(seqFile, "r")
outputList = open(seqFile.partition('.')[0]+'_cleaned.fasta',"w")

header = []
for line in inputList:
    if line.startswith('>'):
        line = line.rstrip('\n')
        line = line.partition('>')
        line = line[2]. partition(' ')
        header.append(line[0])
        outputList.write('>'+str(line[0])+'\n')
    else:
        outputList.write(line)

inputList.close()
outputList.close()
```

### Some Basic Stats and Analysis

#### Number of sequences, Number of Sequences Per Sample

```python
import numpy as np
infile = open('formatSeqs/rousk_ph_gradient_seqs_formated_cleaned.fasta','r')

num_seqs = 0
samples = {}
seq_length = []
for line in infile:
    if line.startswith('>'):
        num_seqs += 1
        line = line.partition('.')
        line = line[0].partition('>')
        if line[2] not in samples.keys():
            samples[line[2]] = 0
        else:
            samples[line[2]] += 1
            
    else:
        seq_length.append(len(line))
infile.close()
```

### Parse Clusters for Representative Sequences

```python
# CD-HIT-EST OTU Parser
# Find each OTU cluster and its represenative sequence
# Assign all fragments by sample ID to respective OTU list
# Specific for Rousk pH Gradient Dataset

# OUTPUT FORMAT
# Cluster-ID Rep_Seq count_for_Sample1 count_for_Sample2 ... 

inputFile = open('clustering/CLUSTER_SEQ_MAPPING','r')
outputFile = open('clustering/OTU_SAMPLE_MAP.txt','w')
unique_samples = ['PH1', 'PH2', 'PH3', 'PH4', 'PH5', 'PH6', 'PH7', 'PH8', 'PH9', 'PH10','PH11', 'PH12', 'PH13', 'PH14', 'PH15', 'PH16', 'PH17', 'PH18', 'PH19', 'PH20', 'PH21', 'PH22', 'PH23', 'PH24', 'PH25', 'PH26', 'PH27']

D = {}
for line in inputFile:
    line = line.rstrip()
    if line.startswith('>'):
        line = line.partition(' ')
        cluster = line[2]
        D[cluster] = [] # create a new list in dict for OTU cluster
    elif '*' in line:
        line = line.partition('>>')
        line = line[2].partition('...')
        D[cluster+'rep'] = line[0] # new key - add cluster rep seq name to dict
        line = line[0].partition('.')
        D[cluster].append(line[0]) # append cluster rep seq sample origin to dict
    else:
        line = line.partition('>>')
        line = line[2].partition('.')
        D[cluster].append(line[0]) # append seq sample origin to dict cluster

for i in range(0,len(D.keys())/2):
        temp = D[str(i)]
        outputFile.write(str(i) + '\t' + D[str(i)+'rep'] + '\t')
        for j in range(0,len(unique_samples)):
            abundance = temp.count(str(unique_samples[j]))
            outputFile.write(str(abundance) + '\t')    
        outputFile.write('\n')

inputFile.close()
outputFile.close()
```

### Denoise Abundance Matrix

```python
CDM = data.transpose() # Sites-by-Species Transpose
sites, species = CDM.shape
print ('Pre-Processing -- Sites: %s, Species: %s' %(sites,species))
remove = np.nonzero(CDM<=3) # Remove any abundance less than or equal to 3
CDM[remove] = 0

# Remove species not present in at least 9 sites
keep_species = []
for i in range(species):
    test = np.nonzero(CDM[:,i])[0];
    if len(test) > 9:
        keep_species.insert(0,i)
CDM = CDM[:,keep_species];

# Remove sites missing species
numpres = []
for i in range(sites):
    numpres.append(len(np.nonzero(CDM[i,:])[0]));

keep_sites = np.nonzero(np.array(numpres)>1)
CDM = CDM[keep_sites]
#CDM = CDM/np.sum(CDM,axis=0)
print ('Post-Processing -- Sites: %s, Species: %s' %(CDM.shape[0],CDM.shape[1]))
```
